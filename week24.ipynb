Top 20 most advertised car models

```python
x=dict(df.make_model.value_counts())
# split dictionary into keys and values
keys = []
values = []
items = x.items()
for item in items:
    keys.append(item[0]), values.append(item[1])
x_label=keys[:20]
y_label=values[:20]
```

```python
plt.subplots(1,1)
sns.barplot(y_label,x_label)
```
