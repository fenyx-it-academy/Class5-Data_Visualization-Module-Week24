```python
import pandas as pd
import numpy as np
import re
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
drive.mount('/content/drive')
import warnings            
warnings.filterwarnings("ignore") 
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/car_data1.csv')
df = data.copy()
```

```python
df['make-model'].value_counts().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10),fontsize=12);#1
```

```python
df['Gears'].value_counts().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10), fontsize=12, color='red');#2
plt.title("Frequencies of the Gear Type",fontsize=20, color="blue")
plt.xlabel("Gears" ,fontsize=20, color='blue')
plt.ylabel("Frequency", rotation=90, fontsize=20, color='blue')
```

```python
df['Fuel_Type'].value_counts().sort_values(ascending=False).plot.bar(figsize=(15,10),color="blue",fontsize=12);#3
plt.title("Frequencies of the Fuel Type",fontsize=20,color="black")
plt.xlabel("Fuel_Type" ,fontsize=20, color="black")
plt.ylabel("Frequency",rotation=90,fontsize=20, color="black")
```

```python
df.groupby('make-model')['price'].mean().sort_values().head(20).plot.barh( figsize=(15,10),color="purple",fontsize=12);#4
plt.title("Average Prices",fontsize=20, color="blue")
plt.xlabel("Price_means" ,fontsize=20, color='green')
plt.ylabel("Brand", rotation=90,fontsize=20, color='green')
```

```python
df.groupby('Fuel_Type')['co2_emissions'].mean().sort_values(ascending=False).plot.bar(figsize=(15,10),color="grey",fontsize=12);#6
plt.title("Average Prices",fontsize=20, color="blue")
plt.xlabel("Fuel_Type" ,fontsize=20, color='green')
plt.ylabel("co2_emissions",rotation=90,fontsize=20, color='green')
```

```python
df.groupby('make-model')['fuel_consumption'].mean().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10),color="red",fontsize=12);#9
plt.title("Average fuel_consumption",fontsize=20, color="blue")
plt.xlabel("make-model" ,fontsize=20, color='green')
plt.ylabel("fuel_consumption",rotation=90,fontsize=20, color='green')
```

```python
df.groupby('make-model')['Power'].mean().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10),color="blue",fontsize=12);#10
plt.title("Average HP",fontsize=20, color="blue")
plt.xlabel("make-model" ,fontsize=20, color='green')
plt.ylabel("Power",rotation=90,fontsize=20, color='green')
```

```python
df.groupby('make-model')['Empty_weight'].mean().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10),color="black",fontsize=12);#11
plt.title("Average Empty_weight",fontsize=20, color="blue")
plt.xlabel("make-model" ,fontsize=20, color='green')
plt.ylabel("Empty_weight",rotation=90,fontsize=20, color='green')
```
