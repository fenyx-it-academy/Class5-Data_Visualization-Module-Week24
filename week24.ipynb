```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt  
import warnings            
warnings.filterwarnings("ignore") 
```

```python
data = pd.read_csv('car_data_v1.csv')
df = data.copy()
```

# 1-Top 20 most advertised car models

```python
df['make-model'].value_counts().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10),fontsize=12);
```

# 2-Frequencies of the number of vehicle advertisements placed in the advertisement according to the gear type

```python
df['Gears'].value_counts().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10), fontsize=12, color='red');
plt.title("Frequencies of the Gear Type",fontsize=20, color="green")
plt.xlabel("Gears" ,fontsize=20, color='green')
plt.ylabel("Frequency", rotation=65, fontsize=20, color='green')
```

# 3-Frequencies of the number of vehicle advertisements placed in the advertisement according to the fuel type


```python
df['Fuel_Type'].value_counts().sort_values(ascending=False).plot.bar(figsize=(15,10),color="blue",fontsize=12);
plt.title("Frequencies of the Fuel Type",fontsize=20,color="green")
plt.xlabel("Fuel_Type" ,fontsize=20, color="green")
plt.ylabel("Frequency",rotation=65,fontsize=20, color="green")
```

# 4-Graph of average of prices of highest 20 by make-model.

```python
df.groupby('make-model')['price'].mean().sort_values().head(20).plot.barh( figsize=(15,10),color="green",fontsize=12);
plt.title("Average Prices",fontsize=20, color="blue")
plt.xlabel("Price_means" ,fontsize=20, color='green')
plt.ylabel("Brand", rotation=90,fontsize=20, color='green')
```

# 6-CO2 emission graph by fuel type

```python
df.groupby('Fuel_Type')['co2_emissions'].mean().sort_values(ascending=False).plot.bar(figsize=(15,10),color="green",fontsize=12);
plt.title("Average Prices",fontsize=20, color="blue")
plt.xlabel("Fuel_Type" ,fontsize=20, color='green')
plt.ylabel("co2_emissions",rotation=90,fontsize=20, color='green')
```

# 9-average fuel consumption by brands

```python
df.groupby('make-model')['fuel_consumption'].mean().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10),color="green",fontsize=12);
plt.title("Average fuel_consumption",fontsize=20, color="blue")
plt.xlabel("make-model" ,fontsize=20, color='green')
plt.ylabel("fuel_consumption",rotation=90,fontsize=20, color='green')
```

# 10-HP average highest 20 by brand

```python
df.groupby('make-model')['Power'].mean().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10),color="green",fontsize=12);
plt.title("Average HP",fontsize=20, color="blue")
plt.xlabel("make-model" ,fontsize=20, color='green')
plt.ylabel("Power",rotation=90,fontsize=20, color='green')
```

# 11-vehicle weight chart by brand

```python
df.groupby('make-model')['Empty_weight'].mean().sort_values(ascending=False).head(20).plot.bar(figsize=(15,10),color="green",fontsize=12);
plt.title("Average Empty_weight",fontsize=20, color="blue")
plt.xlabel("make-model" ,fontsize=20, color='green')
plt.ylabel("Empty_weight",rotation=90,fontsize=20, color='green')
```
