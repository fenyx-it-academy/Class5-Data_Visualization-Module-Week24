```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/autoscout_data_2000_week24_clean_data.csv") # to import csv data from your Google Drive.

df = data.copy() # to copy your data set.
```

```python
df = df.drop("Unnamed: 0",axis=1)
df = df.drop("Unnamed: 0.1",axis=1)
```

```python
df.head(4)
```

### **1-Top 20 most advertised car models**

```python
df['model'].value_counts().head(20).sort_values().plot.barh(figsize=(15,10),color="darkblue",fontsize=12)
plt.title("Top 20 most advertised car models",fontsize=20,color="darkblue")
plt.xlabel("Frequency" ,fontsize=15)
plt.ylabel("Car Model",rotation=90,fontsize=15)
```

```python

df['model'].value_counts().head(20).plot(kind='barh')
```

### **2-Frequencies of the number of vehicle advertisements placed in the advertisement according to the gear type**

```python
df['gears'].value_counts().sort_values().plot(kind='barh')
```

### **3-Frequencies of the number of vehicle advertisements placed in the advertisement according to the fuel type**

```python
df['fuel_type'].value_counts().plot(kind='barh')
```

### **4-Graph of average prices by brand.**

```python
avgprice = sns.barplot(x='make', y="price", data=df)
avgprice.figure.set_size_inches(20, 10)
```

```python

avg_sal = df.groupby('make', sort=False)['price'].mean()
aaa = avg_sal.plot(kind='bar', rot=0, figsize=(16,4), width=.8)
aaa.set_xlim(-.5, 20.5)
aaa.set_ylabel('Mean Price')
```

```python
df.groupby('make')['price'].mean().plot.barh(figsize=(20,15),color="blue",fontsize=12)
plt.title("average prices by brand. ",fontsize=25,color="red")
plt.xlabel("price" ,fontsize=20)
plt.ylabel("make",rotation=90,fontsize=20)
```

### **5-Graph of mileage averages by law**

```python
ax = sns.barplot(x='make', y="mileage", data=df)
ax.figure.set_size_inches(20, 10)
```

### **6-CO2 emission graph by fuel type**

```python
CO2 = sns.barplot(x='fuel_type', y="co2_emissions", data=df)
CO2.figure.set_size_inches(20, 10)
```

```python
avg_sal = df.groupby('fuel_type', sort=False)['co2_emissions'].mean()
ax = avg_sal.plot(kind='bar', rot=0, figsize=(16,4), width=.8)
ax.set_xlim(-.5, 4.5)
ax.set_ylabel('Mean co2_emissions')
```
