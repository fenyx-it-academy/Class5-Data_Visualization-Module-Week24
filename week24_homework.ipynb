### **1-Top 20 most advertised car models**

```python
df['model'].value_counts().head(20).sort_values().plot.barh(figsize=(15,10),color="blue",fontsize=12)
plt.title("Top 20 most advertised car models",fontsize=20,color="blue")
plt.xlabel("Frequency" ,fontsize=15)
plt.ylabel("Car Model",rotation=90,fontsize=15)
```

### **2-Frequencies of the number of vehicle advertisements placed in the advertisement according to the gear type**

```python
df['model'].value_counts().head(20).sort_values().plot.barh(figsize=(15,15),color="blue",fontsize=12)
plt.title("Top 20 most advertised car models",fontsize=25,color="blue")
plt.xlabel("Frequency" ,fontsize=10)
plt.ylabel("Car Model",rotation=90,fontsize=15)
```

```python
df['gears'].value_counts().sort_values().plot.barh(figsize=(10,5),color="red",fontsize=12)
plt.title("Gear Types Frequency ",fontsize=25,color="blue")
plt.xlabel("Frequency" ,fontsize=10)
plt.ylabel("Gears Type",rotation=90,fontsize=10)
```

### **3-Frequencies of the number of vehicle advertisements placed in the advertisement according to the fuel type**

```python
df['fuel_type'].value_counts().sort_values().plot.barh(figsize=(10,5),color="black",fontsize=12)
plt.title("Fuel Type Frequency ",fontsize=25,color="red")
plt.xlabel("Frequency" ,fontsize=10)
plt.ylabel("Fuel Type",rotation=90,fontsize=10)
```

### **4-Graph of average prices by brand.**

```python
from itertools import groupby


df.groupby('make')['price'].mean().plot.barh(figsize=(20,15),color="blue",fontsize=12) #with blot.parh
plt.title("average prices by brand. ",fontsize=25,color="red")
plt.xlabel("price" ,fontsize=20)
plt.ylabel("make",rotation=90,fontsize=20)
```

```python
ax = sns.barplot(x='make', y="price", data=df) #with sns barplot
ax.figure.set_size_inches(20, 10)
```

### **5-Graph of mileage averages by price**

```python
ax = sns.barplot(x='make', y="mileage", data=df) #with sns barplot
ax.figure.set_size_inches(20, 10)
```

### **6-CO2 emission graph by fuel type**

```python

df.groupby('fuel_type')['co2_emissions'].mean().plot.barh(figsize=(10,5),color="pink",fontsize=12) #with blot.parh
plt.title("CO2 emission graph by fuel type",fontsize=15,color="green")
plt.xlabel("co2_emissions means" ,fontsize=10)
plt.ylabel("fuel type",rotation=90,fontsize=10)
```
